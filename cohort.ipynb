```python
import pandas as pd

file_name = "male_vs_female.csv"
df_cohort = pd.read_csv(file_name)

unique_clusters1 = df_cohort["cluster"].drop_duplicates()


cohort_data = {
    'Cohort_Id': range(1, len(unique_clusters1) + 1),
    'CohortName': unique_clusters1
}

df_cohort_w = pd.DataFrame(cohort_data)
df_cohort_w.to_csv('cohort.csv', index=False)
```

```python

file_name = "gene.csv"
gene_data = {
    'GeneId': range(1, len(df_cohort) + 1),
    'GeneName': df_cohort["gene"]
}
df_gene = pd.DataFrame(gene_data)
df_gene.to_csv('gene.csv', index=False)
```

```python




mapped_df = df_cohort.merge(df_cohort_w, left_on='cluster', right_on='CohortName', how='left')

mapped_df = mapped_df.merge(df_gene, left_on='gene', right_on='GeneName', how='left')
# mapped_df

# mapped_df = mapped_df.dropna()

mapped_df[['Cohort_Id', 'GeneId', 'avg_log2FC']].to_csv('cohort_gene.csv', index=False)
```
